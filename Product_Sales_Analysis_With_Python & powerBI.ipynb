# Analysing sales per brand in a given year

```python
# best year of sales
data['year'] = pd.DatetimeIndex(data['timestamp']).year
data.groupby('year')['rating'].count().plot(kind='bar')
```

```python
# By above gragh we know the to sales year are 2015, 2016 and 2014
# Lets analyse the top brands sold in the above top three years
```

```python
#2015
data_2015 = data[data['year'] == 2015]

data_2015.groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#2016
data_2016 = data[data['year'] == 2016]

data_2016.groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#2014
data_2014 = data[data['year'] == 2016]

data_2014.groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
# we can see that top saleing brand is Bose followed by Logitech in above three year
```

```python
# Year 2015 had the best sales.

# Best month of sales

data['month'] = pd.DatetimeIndex(data['timestamp']).month

data.groupby('month')['rating'].count().plot(kind='bar')
```

```python
# Product by category sold the most?

data.groupby('category')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#  From above we know the top product type sold is Headphones.
```

```python
data.groupby('brand')['rating'].count().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#  From above we know the top Brand sold is Logitech and Bose.
```
